```python
1.下载并运行容器
docker pull uifd/ui-for-docker
docker run -it -d --name docker-web -p 9000:9000 -v /var/run/docker.sock:/var/run/docker.sock docker.io/uifd/ui-for-docker
-对于 A/B 命名方式的镜像：
    A 部分：表示镜像的命名空间（Namespace），通常是镜像的发布者、组织或仓库的名称。它可以是：Docker Hub 上的用户名或组织名：例如 uifd 是发布者的名字。
    也可以是自定义镜像仓库地址：例如 myregistry.com/myrepo。 
    B 部分：表示具体的镜像名称（Repository Name），用于区分同一命名空间下的不同镜像。例如 ui-for-docker 是镜像的名称。
-默认命名空间：
    如果镜像名称中没有明确指定命名空间（即没有 /），Docker 会默认使用 library 作为命名空间。
    例如：docker pull ubuntu 实际上等同于 docker pull library/ubuntu
-镜像标签Tag：
    在 A/B 格式的基础上，还可以通过 :标签名 来指定镜像的版本。
    uifd/ui-for-docker:latest：表示拉取 uifd 发布的 ui-for-docker 镜像的 latest 版本。未注明tag时，默认使用latest标签。
-私有镜像仓库
    如果镜像存储在自定义的镜像仓库中，可以通过 仓库地址/镜像名称 的格式来拉取。
    例如：myregistry.com/myrepo/myimage:tag：表示从 myregistry.com 的 myrepo 仓库中拉取 myimage 镜像的指定标签。
```
